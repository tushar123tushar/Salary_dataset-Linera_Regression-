```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
X_train
```

```python
y_train
```

```python
#StandardScaler

from sklearn.preprocessing import StandardScaler

Standard = StandardScaler()

Standard
```

```python
Standard.fit(X_train , y_train)
```

```python
Standard.transform(X_test)
```

```python
#import Liner regression

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
liner_Regression = LinearRegression()
liner_Regression
```

```python
liner_Regression.fit(X_train , y_train)
```

```python
mse = cross_val_score(liner_Regression , X_train , y_train , scoring='neg_mean_squared_error' , cv=5)
```

```python
mse
```

```python
np.mean(mse)
```

```python
x_pred = liner_Regression.predict(X_test)
```

```python
x_pred
```

```python
sns.displot(x_pred-y_test , kind='kde')
```

```python
#scoring 
from sklearn.metrics import r2_score
```

```python
score = r2_score(x_pred , y_test)
```

```python
score
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test , x_pred , color = 'Blue' , label='Predicted vs Actual')
plt.plot([y.min() , y.max()] , [y.min() , y.max()], color = 'red' , linewidth=2 , label = 'Best Fit Line')
plt.title('Linear Regression Best Fit Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```
